```python
import pandas as pd
import numpy as np
import networkx as nx
import os
import matplotlib.pyplot as plt
from pyvis.network import Network
import seaborn as sns
import pyvis
```

## Part 2

```python
# Import relationship data

relationship_df = pd.read_csv('country_20th_century_relationships.csv', index_col = 0)
```

```python
relationship_df.head()
```

## Part 3

```python
# Createing network object

G = nx.from_pandas_edgelist(relationship_df, 
                            source = "source", 
                            target = "target", 
                            edge_attr = "value", 
                            create_using = nx.Graph())
```

## Part 4

```python
# Creating static graph

h = plt.figure(figsize = (20,20))
pos = nx.kamada_kawai_layout(G)
nx.draw(G, with_labels = True, node_color = 'red', edge_cmap = plt.cm.Blues, pos = pos)
plt.show() 
```

## Part 5

```python
# Define net object

net = Network()
net = Network(notebook = True,width="1400px", height="900px", bgcolor='#222222', font_color='white')

# Define nodes from the G object
node_degree = dict(G.degree)

# Setting up node size attributes
nx.set_node_attributes(G, node_degree, 'size')
net.from_nx(G)
net.show_buttons(filter_ = True)
net.repulsion()
net.show("20th_century.html")
```

## Part 6

```python
# Adding communities with Leiden algorithm

from cdlib import algorithms
coms = algorithms.leiden(G)
```

```python
new_com = coms.to_node_community_map()
```

```python
new_com
```

## Part 7

```python
# Put the output from new_com into a dictionary

dict_com = {k:v[0] for k,v in new_com.items()}
```

```python
dict_com
```

```python
# Rerun network graph with the communities as an argument

nx.set_node_attributes(G, dict_com, 'group')
com_net = Network(notebook = True, width="1400px", height="900px", bgcolor='#222222', font_color='white')
com_net.from_nx(G)
com_net.show_buttons(filter_ = True)
com_net.repulsion()
com_net.show("20th_communities.html")
```

From the looks of this graph, the communities are separated among the 5 most prominent countries within the 20th century. The UK's role in WW2, Germany's role in WW2, the US's role in WW2 and the Cold War, the Soviet's role in WW2 and the Cold War, and Japan's role in WW2. There's also a 6th group involving countries in the Middle East, and South Asia that aren't direcly connected to the big 5.

To me it seems like the communities are grouped less by the historical timeframe, and more so by country influence.

## Part 8

```python
# Degree centrality

degree_dict = nx.degree_centrality(G)
degree_df = pd.DataFrame(degree_dict.items(), columns=['country','centrality'])
```

```python
degree_df.sort_values(by = ['centrality'], ascending=False, inplace = True)
```

```python
# Plot the degree centrality

plt.figure(figsize = (10, 20))
with sns.dark_palette("xkcd:blue", 22):
    sns.barplot(x = "centrality", y = "country",
    saturation = 0.9, data = degree_df).set_title("20th-Century - degree centrality")
```

The top 10 countries on this graph having the most connections makes a lot of sense. They all played major roles during WW2 among other conflicts during the 20th century.

```python
# Closeness centrality

closeness_dict = nx.closeness_centrality(G)
closeness_df = pd.DataFrame(closeness_dict.items(), columns=['country','centrality'])
```

```python
closeness_df.sort_values(by = ['centrality'], ascending=False, inplace = True)
```

```python
plt.figure(figsize = (10, 20))
with sns.dark_palette("xkcd:blue", 22):
    sns.barplot(x = "centrality", y = "country",
    saturation = 0.9, data = closeness_df).set_title("20th-Century - closeness centrality")
```

Once again the top 10 countries of this graph are the same top 10 as the previous graph, with the reason why they're the top 10 also being the same. Then the countries furthest removed from the top 10 find themselves at the bottom of the graph.

```python
# Betweenness centrality

betweenness_dict = nx.betweenness_centrality(G)
betweennes_df = pd.DataFrame(betweenness_dict.items(), columns=['country','centrality'])
```

```python
betweennes_df.sort_values(by = ['centrality'], ascending=False, inplace = True)
```

```python
plt.figure(figsize = (10, 20))
with sns.dark_palette("xkcd:blue", 22):
    sns.barplot(x = "centrality", y = "country",
    saturation = 0.9, data = betweennes_df).set_title("20th-Century - betweenness centrality")
```

The top 5 countries from each previous graph also find themselves at the top of this graph. Throughout both WW2 and The Cold War these 5 countries have gotten involved with many different parts of the world.
