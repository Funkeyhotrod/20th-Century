## Part 2

```python
from textblob import TextBlob 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import nltk
import re
nltk.download('stopwords')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
from collections import Counter
sns.set()
```

## Part 3

```python
data = open('Key_Events_20th_Century_Wiki.txt', encoding='utf-8')
```

```python
# Import txt file

with open('Key_Events_20th_Century_Wiki.txt', 'r', errors='ignore') as file:
    data = file.read().replace('\n', ' ').replace('\t', ' ').replace('â€“', '-')
```

## Part 4

```python
# Sentence tokenization 

from nltk.tokenize import sent_tokenize
tokenized_sent = sent_tokenize(data)
print(tokenized_sent)
```

```python
# Word tokenization

from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(data)
print(tokenized_word)
```

```python
# Create frequency distribution

from nltk.probability import FreqDist
dist_words = FreqDist(tokenized_word)
print(dist_words)
```

```python
# Frequency Distribution Plot

plt.figure(figsize=(8, 3))
dist_words.plot(10,cumulative = False)
plt.show()
```

## Part 5

```python
# Defining stopwords

from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
print(stop_words)
```

```python
# Removing stopwords in words

filtered_words = [] # creates an empty list
for word in tokenized_word:
    if word not in stop_words:
        filtered_words.append(word)
```

```python
filtered_words
```

```python
# Substitute all punctuations marks with a space 

sans_punc = re.sub("[^a-zA-Z]",  # Search for all non-letters
                          " ",          # Replace all non-letters with spaces
                          str(filtered_words))
```

```python
# Word tokenization

tokenized_word_2 = word_tokenize(sans_punc)
print(tokenized_word_2)
```

```python
# Create a new FreqDist

dist_words_filter_2 = FreqDist(tokenized_word_2)
```

```python
# Frequency Distribution Plot

plt.figure(figsize=(8, 3))
dist_words_filter_2.plot(10, cumulative = False)
plt.show()
```

Every sample, exept 'The', from the original plot has been removed.

The frequency counts in this graph don't go up into the thousands.

The graph also declines at a slower rate compared to the first graph.

## Part 6

```python
new_stopwords = ['The', 's', 'th']
```

```python
filtered = []
for word in tokenized_word_2:
    if word not in new_stopwords:
        filtered.append(word)
```

```python
# Adding tags
text = TextBlob(str(filtered))
```

```python
tags_list = text.tags
```

```python
tags_list
```

## Part 7

```python
df_text = pd.DataFrame(tags_list)
df_text.columns = ['Words', "Word type"]
```

```python
df_t = df_text.groupby('Word type').count().reset_index()
```

```python
top10 = df_t.nlargest(10, 'Words')
```

```python
top10
```

## Part 8

```python
plt.figure(figsize = (10, 5))
with sns.dark_palette("xkcd:blue", 10):
    sns.barplot(x = "Words", y = "Word type",
    saturation = 0.9, data = top10).set_title("Key Events 20th Century Wiki - top 10 word types used")
```

## Part 9

```python
# Top 15 Nouns
df = df_text[(df_text['Word type'] == "NN") | (df_text['Word type'] == "NNS") | (df_text['Word type'] == "NNP")]
df.columns = ["Word", "Occurences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurences'], ascending=False)
top15 = y.nlargest(15, 'Occurences')
```

```python
plt.figure(figsize=(15, 5))
with sns.dark_palette("xkcd:blue", 15):
    sns.barplot(x="Word", y="Occurences",
    saturation=0.9, data = top15).set_title("Key Events 20th Century Wiki - most frequently used nouns")
```

While most of these words make sence to be here, I'm very confused why December is the most used noun in this wiki page

```python
# Top 15 Verbs
df = df_text[(df_text['Word type'] == "VB") | (df_text['Word type'] == "VBD")]
df.columns = ["Word", "Occurences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurences'], ascending=False)
top15 = y.nlargest(15, 'Occurences')
```

```python
plt.figure(figsize=(15, 5))
with sns.dark_palette("xkcd:blue", 15):
    sns.barplot(x="Word", y="Occurences",
    saturation=0.9, data = top15).set_title("Key Events 20th Century Wiki - most frequently used verbs")
```

A lot of these words are related to military comands. United being the second most common verb makes sence with its association with the USA. I am very surprized with how much retrieved has compared to the other verbs.

```python
# Top 15 Adjectives
df = df_text[(df_text['Word type'] == "JJ")]
df.columns = ["Word", "Occurences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurences'], ascending=False)
top15 = y.nlargest(15, 'Occurences')
```

```python
plt.figure(figsize=(15, 5))
with sns.dark_palette("xkcd:blue", 15):
    sns.barplot(x="Word", y="Occurences",
    saturation=0.9, data = top15).set_title("Key Events 20th Century Wiki - most frequently used adjectives")
```

Japanese and American being the most used adjectives with Russian also being in the list makes sence, but I don't know why German is not in the top 15. Also Japanese having over 10 more occurences then American is surprizing.

## Part 10

```python
# Creating the total counts for every word in the wiki page
listToStr = ' '.join([str(elem) for elem in filtered])
```

```python
all_counts = Counter(re.sub(r'\W+', ' ', listToStr).split())
```

```python
all_counts
```

```python
# Making the list of countries
countries = pd.read_csv("countries_list_20th_century.csv", index_col = 0)
```

```python
pd.options.display.max_rows = None
```

```python
countries
```

```python
# Making contry frequency counts
coun_list = countries['country_name'].to_list()
```

```python
dict_of_counts = {d : all_counts[d] for d in coun_list}
```

```python
# Changing the list into a dataframe
dct = {v:[k] for v,k in dict_of_counts.items()}  
df = pd.DataFrame(dct)
```

```python
df
```

```python
df = df.transpose().reset_index()
```

```python
df.rename(columns = {"index":"Country", 0:"Times mentioned"}, inplace = True)
```

```python
df
```

## Part 11

```python
plt.figure(figsize=(10, 200))
with sns.dark_palette("#79C", 200):
    sns.barplot(x = "Times mentioned", y = "Country",
    saturation=0.9, data = df.sort_values("Times mentioned", ascending = False)).set_title("Key Events 20th Century Wiki - most frequently mentioned country")
```

The biggest problem with this graph is that the process of Tokenization isolates each word, while there are many countries that have multiple words in their name like South Korea. Also there are multiple ways to adress a country like USA, United States, America, American, etc. This can make counting each time a country is mentioned very difficult to be 100% accurate.

From the graph that I do have, the most active countries during 20th century world events are unexpectedly at the top of this list. Germany, Japan, and States making up the top 3. Considering States is capitalized, I'm confident in saying all of its counts do refer to the US.
